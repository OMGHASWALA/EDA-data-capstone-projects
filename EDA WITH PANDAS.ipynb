```python
median_age = df['Age'].median()
df['Age'].fillna(median_age, inplace=True)#to fill missing values with NA
```

```python
df.drop(['Cabin', 'Embarked'], axis=1, inplace=True)# dropping as the data has too many missing value and not useful
```

```python
df
```

```python
# Calculate the overall survival rate
survival_rate = df['Survived'].mean()
print(f"Survival rate: {survival_rate:.2%}")
```

```python
#creating new column "familysize"
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
```

```python
df
```

```python
df['Alone'] = (df['FamilySize'] == 1).astype(int)#new column "alone" to identify no of passengers traveling alone
df
```

```python
#AgeGroup that categorizes passengers into different age groups
bins = [0, 18, 30, 50, 80]
labels = ['Child', 'Young Adult', 'Adult', 'Senior']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)
```

```python
df['Title'] = df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
df
```

## VISUALIZATION

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x='Sex', y='Survived', data=df)
plt.title("Survival Rate by Sex")
plt.show()
```

```python
sns.boxplot(x='Survived', y='Age', data=df)
plt.title("Distribution of Ages by Survival Status")
plt.show()
```

```python
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title("Relationship between Age and Fare")
plt.show()
```
